```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('../Data/clean_data/matrix_1.csv')
print data.shape
```

```python
print data.info()
```

```python
print data.describe()
```

```python
_,axes = plt.subplots(1,2,figsize = (15,5))
axes[0].hist(data["Origin_Zone"],50)
axes[1].hist(data["Destination_Zone"],50)

plt.show()
```

```python
_,axes = plt.subplots(1,2,figsize = (15,5))
axes[0].hist(data["Start_time"],50)
axes[1].hist(data["End_time"],50)

plt.show()
```

```python
data.groupby(["Subscriber_Class"]).hist(column = ["Origin_Zone","Destination_Zone"],bins = 50,figsize = (15,2.5))
```

```python
data.groupby(["Start_time"]).hist(column = ["Origin_Zone","Destination_Zone"],bins = 50,figsize = (15,2.5))
```
